```python
from scrap_cs_faculty import *
```

```python
org_list = ["CMU-CS","Cornell-CS","MIT-AID","MIT-CS","Stanford-CS","UCB-CS","UIUC-CS",]
```

```python
new_cols = [
    'affiliation', 'interests', 'num_papers', 'num_coauthors', 'citedby',
       'hindex', 'i10index', 'citedby5y', 'hindex5y', 'i10index5y',
    'phd_univ', 'phd_year', 
    'url_homepage', 'url',  'url_author', 
    'research_area',
       'research_concentration', 'research_focus', 
       'job_title', 
    'url_picture', 'img_url',
    'phone', 'email', 'cell_phone', 'office_address', 'department', 'school',
       'scholar_id', 
       'file_author']
```

```python
df_dict = {}
df_1_list = []
df_2_list = []
for org in org_list:
    file_xlsx = f"faculty-{org}.xlsx"
    g_file_xlsx = f"data/cs-faculty-gscholar-{org}.xlsx"
    df_dict[org] = dict(file_xlsx=file_xlsx, g_file_xlsx=g_file_xlsx)
    df_1_list.append(pd.read_excel(Path(file_xlsx)))
    df_2_list.append(pd.read_excel(Path(g_file_xlsx)))
```

```python
df_1 = pd.concat(df_1_list)
df_2 = pd.concat(df_2_list)
```

```python
df_1.set_index("name", inplace=True)
df_2.set_index("name", inplace=True)
```

```python
df_12 = df_1.merge(df_2, on="name", how="left")
```

```python
df_12_new = df_12[new_cols].fillna("")
```
