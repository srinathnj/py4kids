## Treemap


Treemaps display hierarchical data as a set of nested rectangles. Each group is represented by a rectangle, which area is proportional to its value. Using color schemes, it is possible to represent several dimensions: groups, subgroupsâ€¦ 

### Basic Treemap using squarify

https://python-graph-gallery.com/200-basic-treemap-with-python/

```python
import matplotlib.pyplot as plt
import squarify 
```

```python
squarify.plot(sizes=[13,22,35,5], label=["group A", "group B", "group C", "group D"], alpha=.7 )
plt.axis('off')
plt.show()
```

```python
import pandas as pd
df = pd.DataFrame({'nb_people':[8,3,4,2], 'team':["team 1", "team 2", "team 3", "team 4"] })
squarify.plot(sizes=df['nb_people'], label=df['team'], alpha=.6 )
plt.axis('off')
plt.show()
```

```python
#libraries
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import squarify # pip install squarify (algorithm for treemap)&lt;/pre&gt;
 
# Create a dataset:
my_values=[i**2 for i in range(1,10)]
 
# create a color palette, mapped to these values
cmap = matplotlib.cm.Blues
mini=min(my_values)
maxi=max(my_values)
norm = matplotlib.colors.Normalize(vmin=mini, vmax=maxi)
colors = [cmap(norm(value)) for value in my_values]
 
# Change color
squarify.plot(sizes=my_values, label=my_values, color=colors, alpha=.6)
# squarify.plot(sizes=my_values, label=my_values, alpha=.6)
plt.axis('off')
plt.show()
```
