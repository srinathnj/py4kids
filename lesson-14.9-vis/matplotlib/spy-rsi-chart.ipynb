```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from datetime import datetime
```

```python
def plot_rsi(df, etf_ticker="SPY", has_grid=True):
    fig = plt.figure()
    fig.set_figwidth(12)
    fig.set_figheight(8)
    fig.tight_layout()
    ax_wp = fig.add_axes([0.1, 0.5, 0.85, 0.40], xticklabels=[])
    ax_rsi = fig.add_axes([0.1, 0.1, 0.85, 0.40])
    etf_ticker = "SPY"
    last_date = df.iloc[-1].name.strftime("%Y-%m-%d")
    ax_wp.set_title(f"RSI: {last_date}")
    ax_wp.set_ylabel(f"{etf_ticker}")
    df["wp"].plot(ax=ax_wp, color="b", grid=has_grid)

    rsi_max = max(np.nanmax(df['rsi']), np.nanmax(df['rsi_old']))
    rsi_min = min(np.nanmin(df['rsi']), np.nanmin(df['rsi_old']))
    ax_rsi.set_ylabel("RSI / RSI(old)")
    df["rsi"].plot(ax=ax_rsi, color="b", ylim=(rsi_min, rsi_max), grid=has_grid)
    df["rsi_old"].plot(ax=ax_rsi, color="r", style="-.", ylim=(rsi_min, rsi_max), grid=has_grid)
    file_img = f"{etf_ticker}_{last_date}.png"
    fig.savefig(file_img)
```

```python
custom_date_parser = lambda x: datetime.strptime(str(x), "%Y%m%d")
```

```python
file_csv = 'spy-rsi-pandas.csv'
df = pd.read_csv(file_csv, parse_dates=['date_'], date_parser=custom_date_parser)
```

```python
df.set_index("date_", inplace=True)
```

```python
df
```

```python
interval = 340
for i in range(1, df.shape[0], interval):
    n_sp = -i
    n_st = -i - interval
    print(f"{n_st} : {n_sp}")
    try:
        plot_rsi(df[n_st:n_sp])
    except Exception as e: 
        print(str(e))
```
