```python
import pandas as pd
df = pd.DataFrame({'x': [1, 2], 'y': [3, 4], 'z': [5, 6]})

df

# Rename columns
df.columns = ['x', 'y', 'z1']
# Do some operations in place
df['x2'] = df.x * df.x

df
```

```python
import databricks.koalas as ks
kdf = ks.DataFrame({'x': [1, 2], 'y': [3, 4], 'z': [5, 6]})

kdf

# Rename columns
kdf.columns = ['x', 'y', 'z1']
# Do some operations in place
kdf['x2'] = kdf.x * kdf.x

kdf
```

```python
import pandas as pd
import numpy as np
date1 = pd.Series(pd.date_range('2012-1-1 12:00:00', periods=3, freq='M'))
date2 = pd.Series(pd.date_range('2013-3-11 21:45:00', periods=3, freq='W'))
df = pd.DataFrame(dict(Start_date = date1, End_date = date2))
# print(df)

df['End_Start'] = df['End_date'] - df['Start_date']

df['diff_years'] = df['End_Start']/np.timedelta64(1,'Y')
df['diff_months'] = df['End_Start']/np.timedelta64(1,'M')
df['diff_weeks'] = df['End_Start']/np.timedelta64(1,'W')
df['diff_days'] = df['End_Start']/np.timedelta64(1,'D')
df['diff_hours'] = df['End_Start']/np.timedelta64(1,'h')
df['diff_mins'] = df['End_Start']/np.timedelta64(1,'m')
df['diff_seconds'] = df['End_Start']/np.timedelta64(1,'s')
```

```python
df
```

```python
import databricks.koalas as ks
```

```python
kdf = ks.from_pandas(df)
```
