```python
crew_id = ''.join(choice(ascii_lowercase + digits) for i in range(3))
client_id = ''.join(choice(digits) for i in range(10))
teams = ["us-east", "us-west", "us-central"]
lobs = ["retail", "fas"]

vt_meta = [["file_id", "lob", "team", "crew_id", "client_id", "creation_ts"]]
for i in range(100):
    creation_ts = str(datetime.datetime.now())
    lob = choice(lobs)
    team = choice(teams)
    crew_id = 'u'+''.join(choice(ascii_lowercase + digits) for i in range(3))
    client_id = ''.join(choice(digits) for i in range(10))
    vt_meta.append([str(i).zfill(4), lob, team, crew_id, client_id, creation_ts])
```

```python
vt_meta
```

```python
import pandas as pd
```

```python
vt_meta = pd.DataFrame(vt_meta[1:], columns=vt_meta[0])
```
