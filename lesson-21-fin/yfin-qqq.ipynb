```python
# Raw Package
import numpy as np
import pandas as pd

#Data Source
import yfinance as yf
# pip3 install yfinance
# /home/wengong/.local/bin

#Data viz
import plotly.graph_objs as go
```

```python
TICKER = 'QQQ'
data = yf.download(tickers=TICKER, period='5d', interval='1m')
```

```python
data.head()
```

```python
MA15 = 15
MA50 = 50
MA200 = 200

#Draw the middle band, higher band, lowest band
data['MA15'] = data['Close'].rolling(window=MA15).mean()
data['MA50'] = data['Close'].rolling(window=MA50).mean()
data['MA200'] = data['Close'].rolling(window=MA200).mean()
```

```python
#declare figure
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=data.index, 
        y= data['MA15'],
        line=dict(color='blue', width=.7), 
        name = 'MA15'
    )
)
fig.add_trace(
    go.Scatter(
        x=data.index, 
        y= data['MA50'],
        line=dict(color='red', width=1.5), 
        name = 'MA50'
    )
)
fig.add_trace(
    go.Scatter(
        x=data.index, 
        y= data['MA200'],
        line=dict(color='green', width=1.5), 
        name = 'MA200'
    )
)


#Candlestick
fig.add_trace(
    go.Candlestick(
        x=data.index,
        open=data['Open'],
        high=data['High'],
        low=data['Low'],
        close=data['Close'], 
        name = f'{TICKER}'
    )
)

# Add titles
fig.update_layout(
    title=f'{TICKER} quote from yfinance',
    yaxis_title='Price')

# X-Axes
fig.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=5, label="5m", step="minute", stepmode="backward"),
            dict(count=15, label="15m", step="minute", stepmode="backward"),
            dict(count=1, label="HTD", step="hour", stepmode="todate"),
            dict(count=3, label="3h", step="hour", stepmode="backward"),
            dict(step="all")
        ])
    )
)

#Show
fig.show()
```
