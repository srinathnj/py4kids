```python
with open("prediction-points.txt") as f:
    data = [i.strip()  for i in f.read().split("\n") if i.strip() and i.strip()[0] != '#']
```

```python
len(data), data
```

```python
from dateutil.parser import *
import datetime
```

```python
def convert_str2int_list(l):
    l2 = [f"{float(i):.2f}" for i in l]
    return sorted([float(i) for i in l2])
```

```python
columns = ["date", "es-kp", "es-cp", "nq-kp", "nq-cp"]
pp_data = []
for i in range(0, len(data), 7):
    # print(f"date: {data[i]}")
    pp_data.append([datetime.datetime.strftime(parse(data[i]),'%Y-%m-%d'), 
                    convert_str2int_list(data[i+2].replace("KP- ", "").split()), 
                    convert_str2int_list(data[i+3].replace("CP- ", "").split()),
                    convert_str2int_list(data[i+5].replace("KP- ", "").split()), 
                    convert_str2int_list(data[i+6].replace("CP- ", "").split())                  
            ])
```

```python
pp_data
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(pp_data, columns=columns)
```

```python
df
```

```python
df[["date", "es-kp", "es-cp"]].to_csv("es-pp.csv",index=False)
```
