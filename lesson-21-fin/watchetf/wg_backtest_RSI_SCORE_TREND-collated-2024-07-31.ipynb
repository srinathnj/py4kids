```python
def _get_img_path(dt, nday=250):
#     dt = dt.replace("/","")
    curr_dir = "."  # os.getcwd()
    trend_img = f"images/{dt}/Trend/RSI_Score_Trend-{nday}-v4.png"
    spy_img = f"images/{dt}/Trend/SPY-{nday}.png"
    qqq_img = f"images/{dt}/Trend/QQQ-{nday}.png"
    iwm_img = f"images/{dt}/Trend/IWM-{nday}.png"
    return [os.path.join(curr_dir,trend_img), 
            os.path.join(curr_dir,spy_img), 
            os.path.join(curr_dir,qqq_img), 
            os.path.join(curr_dir,iwm_img), ]

def trow(dt, width=800, height=1000):
    ret_err = []
    trend_img, spy_img,  qqq_img, iwm_img = _get_img_path(dt)
    ret = ""

    for img in _get_img_path(dt):
    
        if Path(img).exists():
            ret += f"""
            <tr> 
            <img src='{img}' width={width} height={height} />
            </tr>"""
        else:
            ret_err.append(img)

    return ret, ret_err
```

```python
dt = "20240729"
print(_get_img_path(dt))
```

```python
errors = []
html = "<table>"
for dt in all_dates:
    ret, ret_err = trow(dt)
    if ret_err: 
        errors += ret_err
    if ret:    
        html += ret
html += "</table>"
display(HTML(html))

print(errors)
```
