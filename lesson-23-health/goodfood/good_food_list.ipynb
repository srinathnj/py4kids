```python
import re

import inspect

def varname(var):
    """
    Gets the name of var. Does it from the out most frame inner-wards.
    :param var: variable to get name from.
    :return: string

    https://stackoverflow.com/questions/18425225/getting-the-name-of-a-variable-as-a-string
    """
    callers_local_vars = inspect.currentframe().f_back.f_locals.items()
    names = [var_name for var_name, var_val in callers_local_vars if var_val is var]
    return names[-1] if len(names) > 0 else None 

def prep_food_list(part1, parent_word=None):
    words = [i.strip() for i in part1.split(" ") if i.strip()] # remove space
    food_list = words[:-1]
    if parent_word:
        food_list.append(f"{words[-1]} {parent_word}")
    else:
        food_list.append(words[-1])
    return food_list

def scrap_food_list(food_string):
    food_string = food_string.replace('\n',"").strip()
    food_list = []
    ipos = 0
    pat1 = "\(.*?\)"  # non-greedy
    # m = re.search(pat1, Immunity)
    for m in re.finditer(pat1, food_string):
        # print(m)
        istart = ipos
        istop = m.start()
        part1 = food_string[ipos:istop]
        # print(f"{part1}\n {m.group(0)}\n")
        ipos = m.end() + 1
        food_list += prep_food_list(part1, m.group(0))

    if len(food_string) > (ipos-1):
        # last part
        part1 = food_string[ipos:]
        # print(f"{part1}")
        food_list += prep_food_list(part1)
    return [i.strip() for i in food_list]

```

```python
def merge_food_word(food_list):
    "word starts with lower-case is part of current phrase"
    lst = []
    tmp = []
    i = 0
    len_food_list = len(food_list)
    while i < len_food_list:
        word = food_list[i]
        tmp.append(word)
        # print(f"i={i}, word={word}, tmp={tmp}: lst={lst} ")
        if (i+1) < len_food_list:
            next_word = food_list[i+1]
            if next_word[0].isupper():  # peek into next word
                lst.append(' '.join(tmp))
                tmp = []
            i += 1
            continue
        else: # last word
            if food_list[-1][0].islower():  # peek into next word
                tmp.append(food_list[-1])
            lst.append(' '.join(tmp))
            break

    return lst

# test_lst = ['Treviso', 'Apple', 'peel', 'Tieguanyin',  'Tieguanyin', 'green', 'tea', 'Tuna']
# merge_food_word(test_lst)
```
