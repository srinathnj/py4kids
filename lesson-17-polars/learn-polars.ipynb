```python
import polars as pl

df = pl.read_csv("https://j.mp/iriscsv")
```

```python
df.shape
```

```python
df.columns
```

```python
df
```

```python
print(df.filter(pl.col("sepal_length") > 5)
      .groupby("species", maintain_order=True)
      .agg(pl.all().sum())
)
```

```python
df.filter(pl.col("sepal_length")>5)
```

### Lazy API

```python
(
    df.lazy()
    .filter(pl.col("sepal_length") > 5)
    .groupby("species", maintain_order=True)
    .agg(pl.all().sum())
    .collect()
)
```
