## Read CSV in Polars

```python
# gives dtypes hint for Incident Zip column, otherwise, error is thrown
df_311_pl = pl.read_csv('../data/311-service-requests.csv', encoding='utf-8', sep=',', 
                        dtypes={"Incident Zip":str,
#                                 'Created Date':pl.Datetime, 
#                                 'Closed Date':pl.Datetime, 
#                                 'Due Date':pl.Datetime, 
#                                 'Resolution Action Updated Date':pl.Datetime
                               }, parse_dates=False)
```

```python
df_311_pl.shape, df_311_pl.dtypes, df_311_pl.columns
```

```python
df_311_pl.head(5)
```

```python
df_311_pl.limit(5)
```

```python
pl.Config.set_fmt_str_lengths(50)   # Set the number of characters used to print string values
```

```python
df_311_pl_2 = df_311_pl.limit(50).with_columns([
    pl.col("Created Date").str.strptime(pl.Datetime, "%m/%d/%Y %I:%M:%S %p").alias("created_datetime")
])

df_311_pl_2.select(["Created Date", "created_datetime"])
```
