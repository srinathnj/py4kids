copy reviewed html files to py4kids folder

```python
import pathlib
from shutil import copy
```

```python
tgt_dir = r"C:\Users\w_gon\projects\wgong\py4kids\lesson-16-vis\pyvis\examples"
```

```python
tgt_p = pathlib.Path(tgt_dir)
```

```python
html_files = []
with open("readme.md") as f:
    for l in f.read().split('\n'):
        if l.endswith('.html'):
            html_files.append(l)
```

```python
for f in html_files:
    print(f"cp {f} {tgt_p / pathlib.Path(f).parts[-1]}")
    copy(f, tgt_p / pathlib.Path(f).parts[-1])
```
