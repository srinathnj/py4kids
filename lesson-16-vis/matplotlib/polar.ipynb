```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = np.arange(0, 2, 0.01)
theta = 2 * np.pi * r
fig, ax = plt.subplots(
  subplot_kw = {'projection': 'polar'} 
)
ax.plot(theta, r)
ax.set_rticks([0.5, 1, 1.5, 2])
ax.grid(True)
plt.show()
```

```python

r = np.arange(0, 2 * np.pi, 0.02)
fig, ax = plt.subplots()
ax.plot(np.sin(r))
ax.grid(True)
plt.show()
```
