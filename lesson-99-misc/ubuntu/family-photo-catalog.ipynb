# Catalog family photos from duplicate sources

````python
import os
# from pathlib import PurePath
from os.path import join
from PIL import Image
import pandas as pd
from time import time
````

````python
def get_img_timestamp(img_path):
    try:
        exif = Image.open(img_path).getexif()
        DateTime = exif.get(306)
        DateTimeOriginal = exif.get(36867)
        DateTimeDigitized = exif.get(36868)
        return [DateTime if DateTime else "",
                DateTimeOriginal if DateTimeOriginal else "", 
                DateTimeDigitized if DateTimeDigitized else ""]
    except:
        return ["", "", ""]
    
if False:
    for img_path in ['/mnt/data/zapme/photos/2000-01-04/DSC00002.JPG', 
                 "/mnt/data/zapme/tmp/copy1/photos/20080615/IMG_4377.JPG"]:
        print(get_img_timestamp(img_path))
        

COLUMNS = ["folder","filename","dt_created","dt_original", "dt_digitized", "file_path", "filename_std", "file_type"]
````

````python
FILE_TYPES = ['pdf', 'tif', 'jpg', 'doc', 'png', 'zip', 'bmp', 'pps', 'm4v', 'mp4', 'txt',
 'mpg', 'vsd', 'mov', 'gif', 'wav', 'avi', 'rar', 'jpeg', 'mp3', 'xps', 'dotx', 'dot', 'rtf', 'ppt',
 'svg', 'xlsx', 'xls','htm', 'html', 'thm', 'spj']
````

````python
n = 0
data = []
# photos = set()
debug_flag = False  #  True # 

photo_dirs = ["/mnt/data/zapme/photos", 
              "/mnt/data/zapme/tmp/copy1/photos", 
              "/mnt/data/zapme/tmp/copy2/photos", 
              "/mnt/data/zapme/videos", 
              "/mnt/data/Samsung-Phone/WeChat",
             ]
ts_start = time()
for photo_dir in photo_dirs:
    for root, dirs, files in os.walk(photo_dir):
        print(f"root = {root} ...")
        for f in files:
            fn = f.lower()
            ext = fn.split(".")[-1]
            if ext not in FILE_TYPES:
                continue
            n += 1
            folder = root.split("photos")[-1]
            img_info = [folder, f] + get_img_timestamp(join(root,f)) + [root, fn, ext]
            data.append(img_info)
#             photos.add(fn)
        if debug_flag and n > 50:
            break
            
#79707, 19405,

ts_stop = time()

print(f"Completed collecting photos/videos in {ts_stop - ts_start} sec!")
````

````python
df = pd.DataFrame(data, columns=COLUMNS)
````

````python
import sqlite3
````

````python
class DBConn(object):
    def __init__(self, db_file):
        self.conn = sqlite3.connect(db_file)
    def __enter__(self):
        return self.conn
    def __exit__(self, type, value, traceback):
        self.conn.close()
````

````python
TABLE_NAME = "family_photos"
FILE_DB = f"{TABLE_NAME}.sqlite"
cols_defn = [f"{col} text" for col in COLUMNS] 
tmp = ",\n\t".join(cols_defn)
create_photos_sql = f'''
CREATE TABLE if not exists {TABLE_NAME} ( 
    {tmp} 
);
'''
````

# remove duplicate files

How to update dup_rank: 
https://stackoverflow.com/questions/68961876/sqlite-populate-new-column-that-ranks-the-existing-rows

```
alter table family_photos add COLUMN dup_rank INT;

WITH cte AS (
  SELECT filename_std, dt_created, file_path, length(file_path), ROW_NUMBER() OVER (PARTITION BY filename_std,dt_created ORDER BY length(file_path)) rn
  FROM family_photos
)
UPDATE family_photos as f
SET dup_rank = (SELECT rn FROM cte c WHERE f.filename_std = c.filename_std and 'x'||f.dt_created = 'x'||c.dt_created and f.file_path = c.file_path);

```

see `family_photo.sql` the logic to tag duplicate img files


````python
with DBConn(FILE_DB) as _conn:
    cur = _conn.cursor()
    cur.executescript(create_photos_sql)
    _conn.commit()
    
    df.to_sql(TABLE_NAME, _conn, if_exists='append', index=False)
    _conn.commit()
````

````python
with DBConn(FILE_DB) as _conn:
    sql_stmt = """
    select file_path, filename, folder, dup_rank, dt_created from family_photos 
    where dup_rank > 1
    order by file_path,dt_created
    """
    df_dup = pd.read_sql(sql_stmt, _conn)
````

# copy unique photos into new folder organized by 'YYYYMM'

````python
with DBConn(FILE_DB) as _conn:
    sql_stmt = """
    select file_path, filename, folder, dup_rank, dt_created, dt_original from family_photos 
    where dup_rank = 1
    """
    df1 = pd.read_sql(sql_stmt, _conn)
````

````python
uniq_files = df1.to_dict('records')
````

````python
import shutil
import os
````

````python
DEST_FOLDER = "/mnt/data/family_photos"
for i in uniq_files:
    src_fpath = join(i['file_path'], i['filename'])
    if i['dt_original']:
        parent = i['dt_original']
    else:
        if i['folder']:
            parent = i['folder']
            if parent[0] == '/':
                parent = parent[1:]
        else:
            parent = "MISC"
    dest_fpath = join(DEST_FOLDER, parent, i['filename'])
    parent_dir = os.path.dirname(dest_fpath)
    if not os.path.exists(parent_dir):
        os.makedirs(parent_dir, exist_ok=True)
    shutil.copy(src_fpath, dest_fpath)
    # print(f"shutil.copy({src_fpath}, {dest_fpath})")
````
