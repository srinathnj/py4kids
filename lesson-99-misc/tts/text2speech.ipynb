## pyttsx3 

- https://pypi.org/project/pyttsx3/
- https://pyttsx3.readthedocs.io/en/latest/engine.html

````python
import pyttsx3;
engine = pyttsx3.init()
````

````python
voices = engine.getProperty('voices')
for n,voice in enumerate(voices):
    print(f"Voice #{n}: \n{voice.name} ({voice.id})")

print("""
[0] # male - TTS_MS_EN-US_DAVID_11.0
[1] # female - TTS_MS_EN-US_ZIRA_11.0
[2] # female - Chinese
[3] # female - German
""")
````

````python
# in English
engine.setProperty('voice', voices[0].id) 
engine.setProperty('rate', 200)
engine.setProperty('volume',1.0)
engine.say("I will speak this text");

engine.setProperty('voice', voices[1].id)
engine.setProperty('rate', 125)  # slower than normal rate=200
engine.setProperty('volume',2.0) # louder than normal volume=1.0
engine.say("I will speak this text");

# in Chinese
engine.setProperty('voice', voices[2].id)
engine.setProperty('rate', 125)  # slower than normal rate=200
engine.setProperty('volume',2.0) # louder than normal volume=1.0
engine.say("我会讲这个文本")

# in German
engine.setProperty('voice', voices[3].id)
engine.setProperty('rate', 200)  # slower than normal rate=200
engine.setProperty('volume',1.0) # louder than normal volume=1.0
engine.say("Ich werde diesen Text sprechen")

# save audio file
# engine.save_to_file('Hello World from Zira', 'zira.mp3')

engine.runAndWait() ;
````
