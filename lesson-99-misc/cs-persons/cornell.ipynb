```python
professors = [l.strip() for l in lines if "professor" in l.lower()]
```

```python
phd_schools = []

for p in professors:
    p = p.replace("Ph.D.,", "PhD").replace("Ph.D.", "PhD").replace("PhD.", "PhD")
    school_year = p.split("PhD")[-1]
    school = " ".join(school_year.split(" ")[:-1])
    (
        phd_schools.append(school
                       .replace(",", " ").replace(".", "")  
                       .replace(" of", "").replace(" at ", " ").replace(" - ", " ")
                       .replace('UC Berkeley', "University California Berkeley")
                       .replace('Pennsylvaniasylvania', "Pennsylvania").replace('UPenn', "University Pennsylvania").replace('Univ Penn', "University Pennsylvania")
                       .replace("University", "Univ")
                       .strip())
    )

len(phd_schools), sorted(phd_schools)
```

```python
assistant_profs = []
associate_profs = []
adjunct_profs = []
emeritus_profs = []
other_profs = []

for p in professors:
    if "assistant professor" in p.lower():
        assistant_profs.append(p)
    elif "emeritus" in p.lower():
        emeritus_profs.append(p)
    elif "adjunct professor" in p.lower():
        adjunct_profs.append(p)
    elif "associate professor" in p.lower():
        associate_profs.append(p)
    else:
        other_profs.append(p)
```

```python
print("full_profs: \t",len(other_profs), len(other_profs) == len(set(other_profs)))
print("associate_profs: \t",len(associate_profs), len(associate_profs) == len(set(associate_profs)))
print("assistant_profs: \t",len(assistant_profs), len(assistant_profs) == len(set(assistant_profs)))
print("emeritus_profs: \t",len(emeritus_profs), len(emeritus_profs) == len(set(emeritus_profs)))
print("adjunct_profs: \t",len(adjunct_profs), len(adjunct_profs) == len(set(adjunct_profs)))
```
