```python
from pygments.lexers import get_all_lexers
```

```python
lexers = [item[1] for item in get_all_lexers()]
lang_names = []
err_names = []
for i in lexers:
    if len(i):
        lang_names.append(i[0])
    else:
        err_names.append(i)
```
